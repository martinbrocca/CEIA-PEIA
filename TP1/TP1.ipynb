```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ventas = pd.read_excel('./Datos_primer_TP_20Co2025_i0601.xlsx')
df_ventas
```

```python
ventas = df_ventas.Ventas
len(ventas)
```

## Plot the distributions per year

```python
years = [2021, 2022, 2023]

for year in years:
    df_year = df_ventas[df_ventas['Fecha'].dt.year == year]
    ventas = df_year['Ventas']
    
    plt.figure(figsize=(8, 5))
    sns.ecdfplot(ventas, label=f'{year} ECDF')
    plt.legend()
    plt.title(f'Sales Distribution - {year}')
    plt.show()
    
    plt.figure(figsize=(8, 5))
    sns.histplot(ventas, stat='density', label=f'{year} Histogram')
    plt.legend()
    plt.title(f'Sales Distribution - {year}')
    plt.show()
    
    plt.figure(figsize=(8, 5))
    sns.kdeplot(ventas, label=f'{year} KDE')
    plt.legend()
    plt.title(f'Sales Distribution - {year}')
    plt.show()
```
